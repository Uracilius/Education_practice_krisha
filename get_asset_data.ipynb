```python
import requests 


data_ids_file = 'data_ids.csv'
output_file = 'asset_data.csv'
```

```python
import requests
import bs4

def retrieve_data(id):
    url = f"https://krisha.kz/a/show/{id}"
    response = requests.get(url)
    soup = bs4.BeautifulSoup(response.content, "html.parser")
    
    values = soup.find_all("div", class_="offer__advert-short-info")
    desc = soup.find_all("div", class_="js-description")
    price = soup.find_all("div", class_="offer__price")
    
    cleaned_values = [value.get_text(strip=True) for value in values]

    cleaned_desc = [d.get_text(strip=True).replace('\n', ' ') for d in desc]
    cleaned_price = [p.get_text(strip=True).replace('\xa0', '') for p in price]
    
    # Safely append the price, description, views, and date to cleaned_values
    cleaned_values.append(cleaned_price[0] if cleaned_price else 'N/A')
    cleaned_values.append(cleaned_desc[0] if cleaned_desc else 'N/A')
    cleaned_values.append(id)
    
    return cleaned_values
```

```python
import csv
import random
import time

with open(data_ids_file, mode='r', newline='', encoding='utf-8') as file:
    reader = csv.reader(file)
    ids = [row[0] for row in reader]  

with open(output_file, mode='a', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    
    for id in ids:
        try:
            data = retrieve_data(id)
            writer.writerow(data)
            print(f"Data for ID {id} appended successfully.")
        except Exception as e:
            print(f"Failed to retrieve or append data for ID {id}: {e}")
        
        
        delay = random.uniform(0, 10)
        print(f"Waiting for {delay:.2f} seconds before the next request...")
        time.sleep(delay)

print(f"All data processed and appended to {output_file}.")
```
